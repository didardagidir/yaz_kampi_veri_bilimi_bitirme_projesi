```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Didar\Downloads\bitirme\50_Startups.csv")
print(df.head())
```

```python
x = df["R&D Spend"]
y = df["Profit"]

plt.scatter(x, y)
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.title("R&D Spend and Profit Comparison")
plt.show()
```

```python
# The scatter plot likely shows a strong positive correlation — higher R&D spending generally results in higher profits.
```

```python
x = df["Administration"]
y = df["Profit"]

plt.scatter(x, y)
plt.xlabel("Administration")
plt.ylabel("Profit")
plt.show()
```

```python
# The scatter plot suggests little to no clear relationship between administration spending and profit.
```

```python
mean_profit = df.groupby("State")["Profit"].mean()

mean_profit.plot(kind="bar")
plt.title("Profit Comparision by State")
plt.xlabel("State")
plt.ylabel("Profit")
```

```python
# The bar chart will show variation in average profits by state, Florida having higher average.
```

```python
columns = ["R&D Spend", "Administration", "Marketing Spend"]

plt.boxplot([df[c] for c in columns], tick_labels=columns)
plt.title("Spend Comparision")
plt.show()
```

```python
# The boxplot will reveal different distributions: marketing spend has a wide range, R&D spend shows consistently high values, and administration spend varies moderately.
```
